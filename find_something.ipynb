```python
import pandas as pd
```

```python
df = pd.read_excel("all_training_activity.xlsx")
df
```

```python
# check if have null value
df.isnull()
```

```python
# check how many kinds of values Campain Name fields have
cn_list= list(df['Campaign Name'].unique())
cn_list
```

```python
# check how many kinds of values Training Status fields have
df['Training Status'].unique()
```

```python
# output that  the value of Training Status fields is not "Completed"
df1 = df[df['Training Status']!="Completed"]
df1
```

```python
# merge first name and last name, in order to check how many people is the state of "inCompleted"
df1['Full Name']=df1['First Name']+" "+df1['Last Name']
df1['Full Name']
```

```python
# distinct Full Name
df1['Full Name'].unique()
```

```python
#split each dataframe according to Compaign Name fields
df_tems= [df1[df1["Campaign Name"] ==i] for i in cn_list]
len(df_tems)
```

```python
for i in df_tems:
    i
```

```python
# including [The inside man S1E1 - Social Engineering]
s1 = df_tems[0]
for j in range(1,len(df_tems)):
    s1 = pd.merge(s1, df_tems[j], how='inner', on=['Full Name'])
# delete fullName fields and modify fields that have "_x" "_y"
s1=s1.drop('Full Name',axis=1).rename(columns=lambda x: x.replace("_x","").replace("_y",""))
s1
```

```python
colum_index=[i for i in range(0, s1.shape[1])]
colum_index_group = [colum_index[i:i+7] for i in range(0,len(colum_index),7)]
df_all=[s1.iloc[:,x] for x in colum_index_group]
pd_all = pd.concat(df_all,axis=0)
pd_all
```

```python
# excluding [The inside man S1E1 - Social Engineering]
s2 = df_tems[0]
for j in range(1,len(df_tems)-1):
    s2 = pd.merge(s2, df_tems[j], how='inner', on=['Full Name'])
# delete fullName fields and modify fields that have "_x" "_y"
s2=s2.drop('Full Name',axis=1).rename(columns=lambda x: x.replace("_x","").replace("_y",""))
s2
```

```python
colum_index=[i for i in range(0, s2.shape[1])]
colum_index_group = [colum_index[i:i+7] for i in range(0,len(colum_index),7)]
df_all=[s2.iloc[:,x] for x in colum_index_group]
pd_all = pd.concat(df_all,axis=0)
pd_all
```
